```python
import numpy as np # untuk membuat array matriks
from tabulate import tabulate  # untuk membuat tabel
import sympy as sy # untuk simbol
from termcolor import colored  # memberi warna teks yang akan ditampilkan

def GaussSeidel(matrix, initial, tol):   # mendefinisikan metode Gauss-Seidel
  n = np.shape(matrix)[0]   # mengambil ukuran matriks/banyak variabelnya
  x0 = initial              # inisialisasi tebakan awal sebagai x0
  x = x0.copy()             # inisialisasi x. Menggunakan copy() karena jika x0 berubah, x tidak akan berubah

  iterasi = []                             # list kosong yang akan diisi oleh iterasi
  j_iterasi =  1                           # inisialisasi iterasi ke-sekian untuk dimasukkan pada tabel (iterasi dimulai dr satu)

  for i in range(n):
    b = matrix[i,n]      # konstanta pada SPL
    for j in range(n):   # metode Gauss-Seidel
      if j!=i:
        if j<i:
          b -= matrix[i,j]*x[j]
        else:
          b -= matrix[i,j]*x0[j]
    x[i] = b/matrix[i,i]                    # hasil aproksimasi Gauss-Seidel
  isi = np.insert(x, 0, j_iterasi) # inisialisasi solusi-solusi x dari hasil iterasi untuk dimasukkan ke tabel
  iterasi.append(isi)              # menambahkan solusi-solusi x ke tabel

  while np.linalg.norm(x-x0, np.inf)>=tol:  # mengecek batas toleransi dengan error mutlak
    x0 = x                                  # inisialisasi x0 sebagai x(k-1)
    x = x0.copy()                           # inisialisasi x sebagai x(k) tapi awalnya sama dengan x0

    for i in range(n):   # metode Gauss-Seidel
      b = matrix[i,n]
      for j in range(n):
        if j!=i:
          if j<i:
            b -= matrix[i,j]*x[j]
          else:
            b -= matrix[i,j]*x0[j]
      x[i] = b/matrix[i,i]

    j_iterasi += 1                   # iterasi bergerak dengan bertambah 1 terus
    isi = np.insert(x, 0, j_iterasi) # inisialisasi solusi-solusi x dari hasil iterasi untuk dimasukkan ke tabel
    iterasi.append(isi)              # menambahkan solusi-solusi x ke tabel

  print(colored('Tabel iterasi Gauss-Seidel:','blue'))  # menampilkan tulisan untuk tabel iterasi Gauss-Seidel
  print(tabulate(iterasi, headers =["Iterasi","X1",'X2','X3'], tablefmt = "fancy_grid")) # menampilkan seluruh hasil x sesuai iterasinya dalam bentuk tabel

  return x        # nilai yang di-return fungsi ini adalah x(k)

def Jacobi(matrix, initial, tol):   # mendefinisikan metode Jacobi
  n = np.shape(matrix)[0]  # mengambil ukuran matriks/banyak variabelnya
  x0 = initial             # inisialisasi tebakan awal sebagai x0
  x = x0.copy()            # inisialisasi x. Menggunakan copy() karena jika x0 berubah, x tidak akan berubah

  iterasi = []                             # list kosong yang akan diisi oleh iterasi
  j_iterasi =  1                           # inisialisasi iterasi ke-sekian untuk dimasukkan pada tabel (iterasi dimulai dr satu)

  for i in range(n):
    b = matrix[i,n]      # konstanta pada SPL
    for j in range(n):   # metode Jacobi
      if j!=i:
        b -= matrix[i,j]*x0[j]
    x[i] = b/matrix[i,i]           # hasil aproksimasi Jacobi
  isi = np.insert(x, 0, j_iterasi) # inisialisasi solusi-solusi x dari hasil iterasi untuk dimasukkan ke tabel
  iterasi.append(isi)              # menambahkan solusi-solusi x ke tabel

  while np.linalg.norm(x-x0, np.inf)>=tol:   # mengecek batas toleransi dengan error mutlak
    x0 = x                                   # inisialisasi x0 sebagai x(k-1)
    x = x0.copy()                            # inisialisasi x sebagai x(k) tapi awalnya sama dengan x0

    for i in range(n):     # metode Jacobi
      b = matrix[i,n]
      for j in range(n):
        if j!=i:
          b -= matrix[i,j]*x0[j]
      x[i] = b/matrix[i,i]

    j_iterasi += 1                   # iterasi bergerak dengan bertambah 1 terus
    isi = np.insert(x, 0, j_iterasi) # inisialisasi solusi-solusi x dari hasil iterasi untuk dimasukkan ke tabel
    iterasi.append(isi)              # menambahkan solusi-solusi x ke tabel

  print(colored('Tabel iterasi Jacobi:','blue'))  # menampilkan tulisan untuk tabel iterasi Jacobi
  print(tabulate(iterasi, headers =["Iterasi","X1",'X2','X3'], tablefmt = "fancy_grid")) # menampilkan seluruh hasil x sesuai iterasinya dalam bentuk tabel

  return x   # nilai x yang direturn fungsi ini adalah x(k)


print('Selamat datang di program iterasi solusi SPL dengan metode Jacobi dan Gauss-Seidel.')
aug_matrix = np.array(eval(input('Masukkan augmented matrix dari SPL yang akan diselesaikan: ')))  # meminta input matriks dari user lalu dibentuk menjadi array
initial_solution = np.array(eval(input('Masukkan tebakan awal solusi: ')), dtype = float)  # meminta input inisiasi awal solusi dengan tipe float
tolerance = eval(input('Masukkan toleransi: '))  # meminta input batas toleransi yang diinginkan
print('Augmented matrixnya adalah:', '\n', aug_matrix, '\n')  # menampilkan array matriks yang terbentuk
print('Aproksimasi solusi SPL tersebut dengan metode Jacobi adalah: {}'.format(Jacobi(aug_matrix, initial_solution, tolerance)),'\n')        # menampilkan solusi x dari metode Jacobi dengan memanggil fungsi Jacobi
print('Aproksimasi solusi SPL tersebut dengan metode Gauss-Seidel adalah: {}'.format(GaussSeidel(aug_matrix, initial_solution, tolerance)), '\n')  # menampilkan solusi x dari metode Gauss-Seidel dengan memanggil fungsi Gauss-Seidel
print('Terima kasih telah menggunakan program ini.') # penutup
```
