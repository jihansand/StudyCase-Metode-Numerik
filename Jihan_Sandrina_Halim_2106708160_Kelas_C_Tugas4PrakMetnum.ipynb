```python
import numpy as np # untuk membuat array matriks
from termcolor import colored  # memberi warna teks yang akan ditampilkan

def LUFactorization(matrix): # mendefinisikan metode LU Factorization
  n = np.shape(matrix)[0] # mengambil ukuran baris dari matriks
  L = np.identity(n) # mendefinisikan L sebagai matriks identitas nxn

  #operasi baris elementer
  for i in range(n):
    for j in range(i+1, n):
      m = matrix[j,i]/matrix[i,i]
      L[j,i] = m   # memasang elemen L_ji menjadi multiplisitas m = a_ji/a_ii
      matrix[j]= matrix[j]-m*matrix[i]
  return (L, matrix)  # mengembalikan ouput berupa matriks L dan matriks yg telah diolah

print('Selamat datang di program faktorisasi LU matriks.')
matrix = np.array(eval(input('Masukkan matriks yang akan di faktorisasi: ')))  # meminta input matriks yg akan difaktorisasi user
print('=================================== \n')  # pemisah

for i in range(10): # loop untuk mengulang operasi sampai 10 kali
  print('Faktorisasi LU matriks A{}'.format(i)) # menampilkan penjelasan matriks berapa yg sedang difaktorisasi
  L = np.round(LUFactorization(matrix)[0], 0) # mengambil L pada LUFactorization, dibulatkan dengan np.round
  U = np.round(LUFactorization(matrix)[1], 0) # mengambil matrix pada LUFactorization, dibulatkan dengan np.round
  A1 = np.round(L + U)  # mendefinisikan A1 sebagai pertambahan dr L dan U
  print('Matriks L \n{0} \nMatriks U \n{1}'.format(L, U)) # menampilkan matriks L dan U yg telah diolah
  matrix = A1 # mengembalikan A1 sebagai A untuk output terakhir
  print(colored('Matriks A{}'.format(i+1), 'blue')) # menampilkan penjelasan matriks A
  print('{}'.format(A1)) # menampilkan matriks A
  print('=================================== \n') # pemisah

print('Diperoleh matriks A10 adalah \n{}'.format(matrix))  # menampilkan hasil
print('\nTerima kasih telah menggunakan program ini.')  # penutup
```
