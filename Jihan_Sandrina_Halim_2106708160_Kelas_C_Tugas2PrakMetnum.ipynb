```python
#import module yang dibutuhkan

import numpy as np  #untuk list
from tabulate import tabulate  #untuk membuat tabel
import matplotlib.pyplot as plt  #untuk membuat plot
import sympy as sy #untuk simbol
from sympy import simplify #menyederhanakan polinomial

#definisi-definisi fungsi

def LagrangePol(x,x_points,y_points):   #mendefinisikan metode lagrange
  pol = 0 #inisiasi nilai awal polinom adalah 0 krn blm jalanin apa2 biar gangaruh ke pertambahan
  n = len(x_points) #cari panjang titik x
  for k in range(n):
    L = 1 #untuk inisiasi nilai L menjadi 1 biar gangaruh ke perkalian
    for i in range(n):
      if i!=k:  #syarat L
        L*=((x-x_points[i])/(x_points[k]-x_points[i]))  #rumus mencari nilai L, sumasi dari perkalian
    pol += y_points[k]*L   #rumus mencari nilai p(x)
  return pol

def DDTableGenerator(x_points,y_points): #mendefinisikan fungsi untuk membuat tabel DD
  DDTable = [y_points] #kolom-kolom pada tabel. Kolom pertama berisi f
  for column in range(1,len(y_points)):
    DDcolumn = []   #list kosong untuk diisi
    for row in range(len(DDTable[-1])-1):  #mulai mengisi kolom tabel
      DD = (DDTable[-1][row+1]-DDTable[-1][row])/(x_points[column+row]-x_points[row])  #rumus untuk aproksimasinya
      DDcolumn.append(DD)
    DDTable.append(DDcolumn)  #tambah kolom yang telah diisi ke tabel
  return DDTable

def ForwardDD(x,x_points,y_points):  #mendefinisikan metode forward NDD
  DDTable = DDTableGenerator(x_points,y_points)
  pol = DDTable[0][0]     #inisiasi kolom pertama baris pertama
  mult_term = 1         #variabel untuk menyimpan nilai (x-x0)(x-x(n-1))
  for k in range(1,len(DDTable)):
    mult_term*=(x-x_points[k-1])      #menghitung (x-x0)(x-x1)...(x-x(n-1))
    pol+=DDTable[k][0]*mult_term      #menghitung nilai interpolasi
  return pol

def BackwardDD(x,x_points,y_points):  #mendefinisikan metode backward NDD
  DDTable = DDTableGenerator(x_points,y_points)
  pol = DDTable[0][-1]    #nilai dari polinom. Inisiasi suku pertama po
  mult_term = 1       #variabel untuk menyimpan nilai dari (x-xn)(x-x(n-1))
  for k in range(1,len(DDTable)):
    mult_term*=(x-x_points[-k])     #menghitung (x-xn)(x-x(n-1))...(x-x1)
    pol+=DDTable[k][-1]*mult_term     #menghitung nilai interpolasi
  return pol

def f(x):   #mendefinisikan untuk input fungsi
  return eval(formula)

def menu():  #mendefinisikan menu untuk pembuka program
  print('Selamat datang di program aproksimasi nilai fungsi.') #print ucapan

menu()  #memanggil fungsi menu()
o = 1

while True:  #menggunakan while true agar dapat diulang terus menerus dan berhenti jika dan hanya jika diperintah untuk berhenti (break)
  opsi1 = sy.Symbol(input('Apakah anda akan melakukan aproksimasi? (Y/N): '))  # nilai y atau n nya dijadikan simbol agar dapat terbaca
  if opsi1 == sy.Symbol('y') or opsi1 == sy.Symbol('Y'):  #jika pengguna ingin mengaproksimasi
    opsi2 = int(input('Silahkan pilih jenis input yang akan dimasukkan. (1.List, 2.Fungsi): '))  #memilih input (tipe integer) yang akan dimasukkan
    if opsi2 == 1:  #jika inputnya titik
      print('Masukkan titik x dan nilai fungsinya.')
      print('Note: Jumlah titik dan nilai yang dimasukkan harus sama')
      titik_x = eval(input('Masukkan list titik-titik berurutan: '))   #input nilai x
      titik_y = eval(input('Masukkan list nilai fungsi dari titik-titik tsb: '))   #input nilai y
      eval_x = eval(input('Masukkan nilai x yang akan diaproksimasi nilai fungsinya : '))   #input nilai x yang akan diaproksimasi
    elif opsi2 == 2:  #jika inputnya fungsi
      formula = input('Masukkan fungsi: ')  #input fungsi
      bawah = eval(input('Masukkan batas bawah: '))   #input batas bawah interval
      atas = eval(input('Masukkan batas atas: '))   #input batas atas interval
      part = eval(input('Masukkan jumlah partisi: '))   #input jumlah partisi yang diinginkan
      eval_x = eval(input('Masukkan nilai x yang akan diaproksimasi nilai fungsinya: '))  #input nilai x yang akan diaproksimasi
      list_x = np.linspace(bawah, atas, part)  #membuat list nilai x dengan rentang dari interval sebanyak partisinya
      list_y = []   #list kosong untuk diisi nilai y yang akan dihitung
      for i in range(len(list_x)):   #menghitung nilai y sebanyak nilai x
        list_y.append(f(list_x[i]))  #menambahkan nilai y yang dihitung ke list


    metode = int(input('Pilih metode yang akan digunakan dalam mengaproksimasi. (1.Lagrange, 2.NDD): ')) #memberi pilihan metode lagrange(1) atau NDD(2) dg input integer
    if metode == 1 and opsi2 == 1: #jika dipilih metode lagrange dengan input titik
      x = sy.Symbol('x')  #x menjadi simbol agar menjadi polinomial bukan integer
      y_aproks = LagrangePol(eval_x,titik_x,titik_y) #memanggil metode lagrange untuk mendapat hasil y dari input yg diberikan dengan inisialisasi
      print('Aproksimasi nilai fungsi di x = {0} adalah y = {1:.5f}'.format(eval_x,y_aproks))  #menampilkan hasil aproksimasi
      hasil_y = str(simplify(LagrangePol(x,titik_x,titik_y)))  #membuat polinomial lagrange dengan nilai x nya adalah simbol x, lalu polinom disederhanakan dg simplify dan diubah menjadi string agar terbaca di plot
      print('Polinomial Lagrangenya adalah = {0}'.format(hasil_y))  #menampilkan polinomial yang sudah dicari

    elif metode == 1 and opsi2 == 2:   #jika dipilih metode lagrange dengan input fungsi
      x = sy.Symbol('x')   #x menjadi simbol agar menjadi polinomial bukan integer
      y_aproks = LagrangePol(eval_x, list_x, list_y)   #memanggil metode lagrange untuk mendapat hasil y dari input yg diberikan dengan inisialisasi
      print('Aproksimasi nilai fungsi di x = {0} adalah y = {1:.5f}'.format(eval_x, y_aproks))  #menampilkan hasil aproksimasi
      hasil_y = str(simplify(LagrangePol(x, list_x, list_y)))    #membuat polinomial lagrange dengan nilai x nya adalah simbol x, lalu polinom disederhanakan dg simplify dan diubah menjadi string agar terbaca di plot
      print('Polinomial Lagrangenya adalah: {0}'.format(hasil_y))   #menampilkan polinomial yang sudah dicari


    elif metode == 2 and opsi2 == 1:   #jika dipilih metode NDD dengan input titik
      DDTable = DDTableGenerator(titik_x,titik_y)  #inisialisasi DDTableGenerator agar x terdeteksi
      opsi3 = int(input('Forward = 1, Backward = 2: '))  #untuk memilih metode NDD forward(1) atau backward(2) dg input integer
      opsi_tabel = sy.Symbol(input('Apakah anda ingin output tabel? (Y/N): ')) #pengguna akan memilih untuk menampilkan tabel atau tidak dengan input symbol y atau n

      for d in range (o):
        if opsi_tabel == sy.Symbol('y') or opsi_tabel == sy.Symbol('Y'):  #jika pengguna ingin output tabel NDD
          def DDTableNew(DDTable, x_points):  #buat fungsi untuk membuat tabel DD
            table_ = []  #list kosong yg akan diisi hasil untuk membentuk tabel
            for e in range(len(DDTable)):  #nambahin definisi DDTableGenerator yang sebelumnya
              e_table = []  #kolom kosong
              for f in range(len(DDTable[e])):
                e_table.append(DDTable[f][e])
              table_.append(e_table)
            for g in range(len(table_)):  #loop untuk menambahkan elemen x_points ke dalam tabel dg insert
              table_[g].insert(0, x_points[g])
            return table_
          table_new = DDTableNew(DDTableGenerator(titik_x,titik_y),titik_x)  #inisialisasi DDTable yg akan diprint
          print(tabulate(table_new, headers = ['x', 'y', 'NDD1', 'NDD2', 'NDD3', 'NDD4', 'NDD5', 'NDD6'], tablefmt = 'pretty'))  #membuat tabel dengan headers dan jenis sbg berikut
        elif opsi_tabel == sy.Symbol('n') or opsi_tabel == sy.Symbol('N'):  #jika pengguna tidak ingin output tabel NDD
          continue  #lanjut keluar looping

      for h in range(o):
        if opsi3 == 1:  #jika pengguna memilih metode forward NDD
          x = sy.Symbol('x')   #x menjadi simbol agar menjadi polinomial bukan integer
          hasil_y = str(simplify(ForwardDD(x,titik_x,titik_y)))   #membuat polinomial forward NDD dengan nilai x nya adalah simbol x, lalu polinom disederhanakan dg simplify dan diubah menjadi string agar terbaca di plot
          print('NDD Polinomialnya: {0}'.format(hasil_y))  #menampilkan polinomial yang sudah dicari
          y_aproks = ForwardDD(eval_x,titik_x,titik_y)  #memanggil metode forward NDD untuk mendapat hasil y dengan inisialisasi
          print("Untuk x = {0}, nilai aproksimasi dengan metode Forward NDD = {1:.5f}".format(eval_x, y_aproks))  #menampilkan hasil aproksimasi
        elif opsi3 == 2:  #jika pengguna memilih metode backward NDD
          x = sy.Symbol('x')   #x menjadi simbol agar menjadi polinomial bukan integer
          hasil_y = str(simplify(BackwardDD(x,titik_x,titik_y)))  #membuat polinomial backward NDD dengan nilai x nya adalah simbol x, lalu polinom disederhanakan dg simplify dan diubah menjadi string agar terbaca di plot
          print('NDD Polinomialnya: {0}'.format(hasil_y))   #menampilkan polinomial yang sudah dicari
          y_aproks = BackwardDD(eval_x,titik_x,titik_y)   #memanggil metode backward NDD untuk mendapat hasil y dengan inisialisasi
          print("Untuk x = {0}, nilai aproksimasi dengan metode Backward NDD = {1:.5f}".format(eval_x, y_aproks))   #menampilkan hasil aproksimasi


    elif metode == 2 and opsi2 == 2:  #jika dipilih metode NDD dengan input fungsi
      DDTable = DDTableGenerator(list_x,list_y)   #penjelasan sama seperti di atas, namun di sini menggunakan adalah list x dan list y karena input berupa fungsi
      opsi3 = int(input('Forward = 1, Backward = 2: '))  #input metode NDD yg akan dipilih
      opsi_tabel = sy.Symbol(input('Apakah anda ingin output tabel? (Y/N): '))  ##pengguna akan memilih untuk menampilkan tabel atau tidak dengan input symbol y atau n

      for d in range (o):
        if opsi_tabel == sy.Symbol('y') or opsi_tabel == sy.Symbol('Y'): #jika memilih iya
          def DDTableNew(DDTable, x_points):
            table_ = []
            for e in range(len(DDTable)):
              e_table = []
              for f in range(len(DDTable[e])):
                e_table.append(DDTable[f][e])
              table_.append(e_table)
            for g in range(len(table_)):
              table_[g].insert(0, x_points[g])
            return table_
          table_new = DDTableNew(DDTableGenerator(list_x,list_y),list_x)
          print(tabulate(table_new, headers = ['x', 'y', 'NDD1', 'NDD2', 'NDD3', 'NDD4', 'NDD5', 'NDD6'], tablefmt = 'pretty'))
        elif opsi_tabel == sy.Symbol('n') or opsi_tabel == sy.Symbol('N'): #jika memilih tidak
          continue  #keluar dr looping

      for h in range(o):
        if opsi3 == 1:  #jika memilih metode forward NDD
          x = sy.Symbol('x')  #x menjadi simbol agar menjadi polinomial bukan integer
          hasil_y = str(simplify(ForwardDD(x,list_x,list_y)))
          print('NDD Polinomialnya: {0}'.format(hasil_y))
          y_aproks = ForwardDD(eval_x,list_x,list_y)
          print("Untuk x = {0}, nilai aproksimasi dengan metode Forward NDD = {1:.5f}".format(eval_x, y_aproks))
        elif opsi3 == 2:  #jika memilih metode backward NDD
          x = sy.Symbol('x') #x menjadi simbol agar menjadi polinomial bukan integer
          hasil_y = str(simplify(BackwardDD(x,list_x,list_y)))
          print('NDD Polinomialnya: {0}'.format(hasil_y))
          y_aproks = BackwardDD(eval_x,list_x,list_y)
          print("Untuk x = {0}, nilai aproksimasi dengan metode Backward NDD = {1:.5f}".format(eval_x, y_aproks))


    show = int(input('Tunjukkan 1.Tabel atau 2.Plot: '))  #pengguna memilih visualisasi yg diinginkan
    if show == 1 and opsi2 == 1:  #memilih tabel dengan input titik
      x_ = np.append(titik_x, eval_x)  #menggabungkan titik x dan eval x nya
      y_ = np.append(titik_y, y_aproks)  #menggabungkan titik y dan titik hasil aproksimasinya
      kolom_x = np.sort(x_)  #mengurutkan nilai x
      kolom_y = np.sort(y_)  #mengurutkan nilai y
      print('Berikut tabelnya:')  #untuk menampilkan petunjuk
      print(tabulate({'x':kolom_x, 'y':kolom_y}, headers = ['x', 'y'], tablefmt = 'pretty'))  #menampilkan tabel dengan kolom x dan y

    elif show == 1 and opsi2 == 2:  #memilih tabel dengan input fungsi
      x_ = np.append(list_x, eval_x)  #menggabungkan titik x dan eval x nya
      y_ = np.append(list_y, y_aproks)  #menggabungkan titik y dan titik hasil aproksimasinya
      kolom_x = np.sort(x_)  #mengurutkan nilai x
      kolom_y = np.sort(y_)  #mengurutkan nilai y
      print('Berikut tabelnya:')  #untuk menampilkan petunjuk
      print(tabulate({'x':kolom_x, 'y':kolom_y}, headers = ['x', 'y'], tablefmt = 'pretty'))  #menampilkan tabel dengan kolom x dan y

    elif show == 2 and opsi2 == 1:   #memilih plot dengan input titik
      def f(x):  #hasil_y didefinisikan menjadi f(x) lalu di eval agar nilai x bisa disubstitusi
        return eval(hasil_y)
      u = np.linspace(titik_x[0], titik_x[len(titik_x)-1],10)
      plt.plot(u, f(u))  #grafik polinom interpolasi
      plt.plot(eval_x, y_aproks, 'r-o')  #titik interpolasi warna merah
      plt.plot(titik_x, titik_y, 'g-')   #titik-titik x dan y yang membentuk grafik warna hijau
      plt.legend(['Polinom interpolasi', 'Titik yang diinterpolasi', 'Fungsi/Titik input'])  #menamai grafik
      plt.title('Grafik Input dan Polinomial Interpolasi')  #menamai plot
      print('Berikut plotnya: ')  #untuk menampilkan petunjuk
      plt.show()   #menampilkan plotnya

    elif show == 2 and opsi2 == 2:   #memilih plot dengan input fungsi
      def f(x):  #hasil_y didefinisikan menjadi f(x) lalu di eval agar nilai x bisa disubstitusi
        return eval(hasil_y)
      u = np.linspace(list_x[0], list_x[len(list_x)-1], 10)
      plt.plot(u, f(u))  #grafik polinom interpolasi
      plt.plot(eval_x, y_aproks, 'r-o')  #titik interpolasi warna merah
      plt.plot(list_x, list_y, 'g-')  #titik-titik x dan y yang membentuk grafik warna hijau
      plt.legend(['Polinom interpolasi', 'Titik yang diinterpolasi', 'Fungsi/Titik input'])  #menamai grafik
      plt.title('Grafik Input dan Polinomial Interpolasi')  #menamai plot
      print('Berikut plotnya: ')  #untuk menampilkan petunjuk
      plt.show()  #menampilkan plotnya

  elif  opsi1 == sy.Symbol('n') or opsi1 == sy.Symbol('N'):  #jika pengguna tidak ingin lagi mengaproksimasi
    print('Terima kasih telah menggunakan program ini.') #salam penutup
    break  #berhenti
```

```python
def LagrangePol(x,x_points,y_points):
  pol = 0
  n = len(x_points)
  for k in range(n):
    L = 1
    for i in range(n):
      if i!=k:
        L*=((x-x_points[i])/(x_points[k]-x_points[i]))
    pol += y_points[k]*L
  return pol

def DDTableGenerator(x_points,y_points):
  DDTable = [y_points]
  for column in range(1,len(y_points)):
    DDcolumn = []
    for row in range(len(DDTable[-1])-1):
      DD = (DDTable[-1][row+1]-DDTable[-1][row])/(x_points[column+row]-x_points[row])
      DDcolumn.append(DD)
    DDTable.append(DDcolumn)
  return DDTable

def ForwardDD(x,x_points,y_points):
  DDTable = DDTableGenerator(x_points,y_points)
  pol = DDTable[0][0]
  mult_term = 1
  for k in range(1,len(DDTable)):
    mult_term*=(x-x_points[k-1])
    pol+=DDTable[k][0]*mult_term
  return pol

def BackwardDD(x,x_points,y_points):
  DDTable = DDTableGenerator(x_points,y_points)
  pol = DDTable[0][-1]
  mult_term = 1
  for k in range(1,len(DDTable)):
    mult_term*=(x-x_points[-k])
    pol+=DDTable[k][-1]*mult_term
  return pol

def f(x):
  return eval(formula)

def menu():
  print('Selamat datang di program aproksimasi nilai fungsi.')

menu()
o = 1

while True:
  opsi1 = sy.Symbol(input('Apakah anda akan melakukan aproksimasi? (Y/N): '))
  if opsi1 == sy.Symbol('y') or opsi1 == sy.Symbol('Y'):
    opsi2 = int(input('Silahkan pilih jenis input yang akan dimasukkan. (1.List, 2.Fungsi): '))
    if opsi2 == 1:
      print('Masukkan titik x dan nilai fungsinya.')
      print('Note: Jumlah titik dan nilai yang dimasukkan harus sama')
      titik_x = eval(input('Masukkan list titik-titik berurutan: '))
      titik_y = eval(input('Masukkan list nilai fungsi dari titik-titik tsb: '))
      eval_x = eval(input('Masukkan nilai x yang akan diaproksimasi nilai fungsinya : '))
    elif opsi2 == 2:
      formula = input('Masukkan fungsi: ')
      bawah = eval(input('Masukkan batas bawah: '))
      atas = eval(input('Masukkan batas atas: '))
      part = eval(input('Masukkan jumlah partisi: '))
      eval_x = eval(input('Masukkan nilai x yang akan diaproksimasi nilai fungsinya: '))
      list_x = np.linspace(bawah, atas, part)
      list_y = []
      for i in range(len(list_x)):
        list_y.append(f(list_x[i]))

    metode = int(input('Pilih metode yang akan digunakan dalam mengaproksimasi. (1.Lagrange, 2.NDD): '))
    if metode == 1 and opsi2 == 1:
      x = sy.Symbol('x')
      y_aproks = LagrangePol(eval_x,titik_x,titik_y)
      print('Aproksimasi nilai fungsi di x = {0} adalah y = {1:.5f}'.format(eval_x,y_aproks))
      hasil_y = str(simplify(LagrangePol(x,titik_x,titik_y)))
      print('Polinomial Lagrangenya adalah = {0}'.format(hasil_y))

    elif metode == 1 and opsi2 == 2:
      x = sy.Symbol('x')
      y_aproks = LagrangePol(eval_x, list_x, list_y)
      hasil_y = str(simplify(LagrangePol(x, list_x, list_y)))
      print('Polinomial Lagrangenya adalah: {0}'.format(hasil_y))
      print('Aproksimasi nilai fungsi di x = {0} adalah y = {1:.5f}'.format(eval_x, y_aproks))

    elif metode == 2 and opsi2 == 1:
      DDTable = DDTableGenerator(titik_x,titik_y)
      opsi3 = int(input('Forward = 1, Backward = 2: '))
      opsi_tabel = sy.Symbol(input('Apakah anda ingin output tabel? (Y/N): '))

      for d in range (o):
        if opsi_tabel == sy.Symbol('y') or opsi_tabel == sy.Symbol('Y'):
          def DDTableNew(DDTable, x_points):
            table_ = []
            for e in range(len(DDTable)):
              e_table = []
              for f in range(len(DDTable[e])):
                e_table.append(DDTable[f][e])
              table_.append(e_table)
            for g in range(len(table_)):
              table_[g].insert(0, x_points[g])
            return table_
          table_new = DDTableNew(DDTableGenerator(titik_x,titik_y),titik_x)
          print(tabulate(table_new, headers = ['x', 'y', 'NDD1', 'NDD2', 'NDD3', 'NDD4', 'NDD5', 'NDD6'], tablefmt = 'pretty'))
        elif opsi_tabel == sy.Symbol('n') or opsi_tabel == sy.Symbol('N'):
          continue

      for h in range(o):
        if opsi3 == 1:
          x = sy.Symbol('x')
          hasil_y = str(simplify(ForwardDD(x,titik_x,titik_y)))
          print('NDD Polinomialnya: {0}'.format(hasil_y))
          y_aproks = ForwardDD(eval_x,titik_x,titik_y)
          print("Untuk x = {0}, nilai aproksimasi dengan metode Forward NDD = {1:.5f}".format(eval_x, y_aproks))
        elif opsi3 == 2:
          x = sy.Symbol('x')
          hasil_y = str(simplify(BackwardDD(x,titik_x,titik_y)))
          print('NDD Polinomialnya: {0}'.format(hasil_y))
          y_aproks = BackwardDD(eval_x,titik_x,titik_y)
          print("Untuk x = {0}, nilai aproksimasi dengan metode Backward NDD = {1:.5f}".format(eval_x, y_aproks))

    elif metode == 2 and opsi2 == 2:
      DDTable = DDTableGenerator(list_x,list_y)
      opsi3 = int(input('Forward = 1, Backward = 2: '))
      opsi_tabel = sy.Symbol(input('Apakah anda ingin output tabel? (Y/N): '))

      for d in range (o):
        if opsi_tabel == sy.Symbol('y') or opsi_tabel == sy.Symbol('Y'):
          def DDTableNew(DDTable, x_points):
            table_ = []
            for e in range(len(DDTable)):
              e_table = []
              for f in range(len(DDTable[e])):
                e_table.append(DDTable[f][e])
              table_.append(e_table)
            for g in range(len(table_)):
              table_[g].insert(0, x_points[g])
            return table_
          table_new = DDTableNew(DDTableGenerator(list_x,list_y),list_x)
          print(tabulate(table_new, headers = ['x', 'y', 'NDD1', 'NDD2', 'NDD3', 'NDD4', 'NDD5', 'NDD6'], tablefmt = 'pretty'))
        elif opsi_tabel == sy.Symbol('n') or opsi_tabel == sy.Symbol('N'):
          continue

      for h in range(o):
        if opsi3 == 1:
          x = sy.Symbol('x')
          hasil_y = str(simplify(ForwardDD(x,list_x,list_y)))
          print('NDD Polinomialnya: {0}'.format(hasil_y))
          y_aproks = ForwardDD(eval_x,list_x,list_y)
          print("Untuk x = {0}, nilai aproksimasi dengan metode Forward NDD = {1:.5f}".format(eval_x, y_aproks))
        elif opsi3 == 2:
          x = sy.Symbol('x')
          hasil_y = str(simplify(BackwardDD(x,list_x,list_y)))
          print('NDD Polinomialnya: {0}'.format(hasil_y))
          y_aproks = BackwardDD(eval_x,list_x,list_y)
          print("Untuk x = {0}, nilai aproksimasi dengan metode Backward NDD = {1:.5f}".format(eval_x, y_aproks))


    show = int(input('Tunjukkan 1.Tabel atau 2.Plot: '))
    if show == 1 and opsi2 == 1:
      x_ = np.append(titik_x, eval_x)
      y_ = np.append(titik_y, y_aproks)
      kolom_x = np.sort(x_)
      kolom_y = np.sort(y_)
      print('Berikut tabelnya:')
      print(tabulate({'x':kolom_x, 'y':kolom_y}, headers = ['x', 'y'], tablefmt = 'pretty'))
    elif show == 1 and opsi2 == 2:
      x_ = np.append(list_x, eval_x)
      y_ = np.append(list_y, y_aproks)
      kolom_x = np.sort(x_)
      kolom_y = np.sort(y_)
      print('Berikut tabelnya:')
      print(tabulate({'x':kolom_x, 'y':kolom_y}, headers = ['x', 'y'], tablefmt = 'pretty'))
    elif show == 2 and opsi2 == 1:
      def f(x):
        return eval(hasil_y)
      u = np.linspace(titik_x[0], titik_x[len(titik_x)-1],10)
      plt.plot(u, f(u))
      plt.plot(eval_x, y_aproks, 'r-o')
      plt.plot(titik_x, titik_y, 'g-')
      plt.legend(['Polinom interpolasi', 'Titik yang diinterpolasi', 'Fungsi/Titik input'])
      plt.title('Grafik Input dan Polinomial Interpolasi')
      print('Berikut plotnya: ')
      plt.show()
    elif show == 2 and opsi2 == 2:
      def f(x):
        return eval(hasil_y)
      u = np.linspace(list_x[0], list_x[len(list_x)-1], 10)
      plt.plot(u, f(u))
      plt.plot(eval_x, y_aproks, 'r-o')
      plt.plot(list_x, list_y, 'g-')
      plt.legend(['Polinom interpolasi', 'Titik yang diinterpolasi', 'Fungsi/Titik input'])
      plt.title('Grafik Input dan Polinomial Interpolasi')
      print('Berikut plotnya: ')
      plt.show()

  elif  opsi1 == sy.Symbol('n') or opsi1 == sy.Symbol('N'):
    print('Terima kasih telah menggunakan program ini.')
    break
```
